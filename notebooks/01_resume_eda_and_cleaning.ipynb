# 01 Resume EDA and Cleaning

This notebook is used for the initial exploration and cleaning of the Resume Dataset.

Main goals:
- load the raw resume dataset
- inspect dataset structure
- check missing values and duplicates
- analyze resume categories
- identify IT-related resume categories
- clean resume text
- save cleaned resume datasets for later extraction and analysis

## 1. Imports and Settings

```python
import pandas as pd
import numpy as np
import re

from pathlib import Path
```

```python
RAW_RESUME_DIR = Path("../data/raw/resumes")
PROCESSED_RESUME_DIR = Path("../data/processed/resumes")

csv_file='Resume.csv'
```

```python
resume_path=RAW_RESUME_DIR/csv_file
print(resume_path)
```

## 3. Missing Values Check

```python
missing_values=df.isna().sum().sort_values(ascending=False)
```

```python
empty_rows = df[
    df["Resume_str"].isna() |
    (df["Resume_str"].astype(str).str.strip() == "")
]
```

```python
df = df.drop(empty_rows.index).reset_index(drop=True)
```

## 5. Identify Important Columns



```python
df_selected=df.drop(columns=["Resume_html"])
```

## 7. Select Relevant Columns

```python
df_resume_it=df_selected[df_selected['Category']=='INFORMATION-TECHNOLOGY'].copy()
```

```python
df_resume_it
```

## 8. Rename Columns


```python
rename_map = {
    'ID': 'id',
    'Category': "category",
    'Resume_str': "resume_text"
}

df_selected = df_selected.rename(columns=rename_map)

df_selected.head()
```

## 9. Clean Resume Text


```python
def clean_resume_text(text):
    if pd.isna(text):
        return ""
    
    text = str(text)

    text = text.replace("\t", " ")
    text = text.replace("\n", " ")
    text = text.replace("\r", " ")

    text = re.sub(r"\s+", " ", text)

    text = text.strip()
    
    return text
```

```python
df_selected["clean_resume_text"] = df_selected["resume_text"].apply(clean_resume_text)

df_selected.head()
```

```python
print(df_selected.iloc[0]["clean_resume_text"])
```

```python
df_selected["text_length"] = df_selected["clean_resume_text"].str.len()

df_selected[["category", "text_length"]].describe()
```

```python
df_selected.sort_values("text_length").head(10)
```

## 10. Remove Duplicate Resume Texts

```python
duplicates_before = df_selected.duplicated(subset=["clean_resume_text"]).sum()

duplicates_before
```

```python
df_resume_cleaned = df_selected.drop_duplicates(subset=["clean_resume_text"]).reset_index(drop=True)

df_resume_cleaned.shape
```

## 11. Filter IT-Related Resume Categories

```python
it_categories = [
    "INFORMATION-TECHNOLOGY"
]

df_resume_it = df_resume_cleaned[
    df_resume_cleaned["category"].isin(it_categories)
].copy()

df_resume_it.shape
```

## 12. Preview Cleaned IT Resume Examples

```python
if len(df_resume_it) > 0:
    print(df_resume_it.iloc[0]["clean_resume_text"])
else:
    print("No IT-related resumes found. Please check category names.")
```

## 13. Save Cleaned Resume Datasets

```python
cleaned_all_path = PROCESSED_RESUME_DIR / "cleaned_resumes_all.csv"
cleaned_it_path = PROCESSED_RESUME_DIR / "cleaned_resumes_it.csv"

df_resume_cleaned.to_csv(cleaned_all_path, index=False)
df_resume_it.to_csv(cleaned_it_path, index=False)

print(f"Saved all cleaned resumes to: {cleaned_all_path}")
print(f"Saved IT cleaned resumes to: {cleaned_it_path}")
```
